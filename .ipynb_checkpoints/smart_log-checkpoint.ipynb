```python
import re

class LogParser:
    def __init__(self, file):
        self.file = "sample.log"
        self.logs = []

    def parse_logs(self):
        pattern = r"(\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}) (INFO|ERROR|WARNING) (.*)"
        
        try:
            with open(self.file, "r") as f:
                data = f.readlines()

            for line in data:
                match = re.match(pattern, line)
                if match:
                    self.logs.append({
                        "time": match.group(1),
                        "level": match.group(2),
                        "message": match.group(3)
                    })

        except FileNotFoundError:
            print("File not found")
        
        return self.logs
```

```python
class LogAnalyzer:
    def __init__(self, logs):
        self.logs = logs

    def count_levels(self):
        result = {"INFO": 0, "ERROR": 0, "WARNING": 0}

        for log in self.logs:
            result[log["level"]] += 1

        return result

    def find_errors(self):
        return [log for log in self.logs if log["level"] == "ERROR"]
```

```python
class ReportGenerator:
    def __init__(self, analysis):
        self.analysis = analysis

    def generate_report(self):
        print("\n--- Log Summary ---")
        for key, value in self.analysis.items():
            print(f"{key}: {value}")
```

```python
parser = LogParser("sample.log")
logs = parser.parse_logs()

analyzer = LogAnalyzer(logs)
summary = analyzer.count_levels()

report = ReportGenerator(summary)
report.generate_report()
```
